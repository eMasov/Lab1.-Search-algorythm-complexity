```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Task 1

```python
data = pd.read_csv("time_record.csv")
data
```

```python
k, b = np.polyfit(data['N'], data['linear_searh_worst'], 1)
x = np.linspace(0, 1000000)


plt.plot(x, k * x + b, label='$T = 0.17 \cdot N$', color='r')
plt.scatter(data['N'], data['linear_searh_worst'], label='linear search 1')
plt.xlabel('N, array length')
plt.ylabel('Time, microseconds')
plt.legend()
plt.title("Linear search, elem is not in array")
plt.show()
```

```python
k, b = np.polyfit(data['N'], data['linear_searh_worst'], 1)
x = np.linspace(0, 1000000)

plt.plot(x, k * x + b, label='$T = 0.17 N$', color='r')
plt.scatter(data['N'], data['linear_searh_average'], label='linear search 2')
plt.xlabel('N, array length')
plt.ylabel('Time, microseconds')
plt.legend()
plt.title("Linear search, elem is in array")
plt.show()
```

```python
k, b = np.polyfit(np.log2(data['N']), data['binsearch_worst'], 1)
k, b
```

```python
k, b = np.polyfit(np.log2(data['N']), data['binsearch_worst'], 1)
x = np.linspace(1, 1000000)
plt.plot(x, k * np.log2(x) + b, label='$T = 0.27 \cdot ln(N) + 4.50$', color='r')
plt.scatter(data['N'], data['binsearch_worst'], label='binary search 1')
plt.xlabel('N, array length')
plt.ylabel('Time, microseconds')
plt.legend()
plt.title("Binary search, elem is not in array")
plt.show()
```

```python
k, b = np.polyfit(np.log2(data['N']), data['binsearch_average'], 1)
k, b
```

```python
k, b = np.polyfit(np.log2(data['N']), data['binsearch_average'], 1)
x = np.linspace(1, 1000000)
plt.plot(x, k * np.log2(x) + b, label='$T = 0.31 \cdot ln(N) + 3.15$', color='r')
plt.scatter(data['N'], data['binsearch_average'], label='binary search 2')
plt.xlabel('N, array length')
plt.ylabel('Time, microseconds')
plt.legend()
plt.title("Binary search, elem is in array")
plt.show()
```

Task 2

```python
data1 = pd.read_csv("task2.1.csv")
data1
```

```python
data2 = pd.read_csv("task2.2.csv")
data2
```

```python
k, b = np.polyfit(data1['N'], data1['time_lin'], 1)
x = np.linspace(1, 1000000)
plt.plot(x, k * x + b, label='$T = 0.25 \cdot N - 903.34$', color='r')
plt.scatter(data1['N'], data1['time_lin'], label='Sum of two, linear')
plt.xlabel('N, array length')
plt.ylabel('Time, microseconds')
plt.legend()
plt.show()
print(k, b)
```

```python
a, b, c = np.polyfit(data2['N'], data2['time_sqr'], 2)
x = np.linspace(1, 5000)
plt.plot(x, a * x ** 2 + b * x + c, label='$T = 0.12 \cdot N^2 + 241.80 \cdot N -162935.20$', color='r')
plt.scatter(data2['N'], data2['time_sqr'], label='Sum of two, square')
plt.xlabel('N, array length')
plt.ylabel('Time, microseconds')
plt.legend()
plt.show()
```
